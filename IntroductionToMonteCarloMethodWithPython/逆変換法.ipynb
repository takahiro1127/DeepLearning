```python
nbins = 50
scale = 1.0
np.random.seed()
N = 100000
U = scipy.stats.uniform(loc = 0.0, scale= 1.0).rvs(size = N)

X1 = -scale * np.log(1 - U)

plt.figure(1)
plt.hist(U, nbins, normed = True)

plt.figure(2)
rv = scipy.stats.expon(scale = scale)
plt.hist(X1, nbins, normed = True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw = 2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))

plt.show()
```

```python
nbins = 50
x0 = 0
gamma = 1
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = x0 + gamma * np.tan(np.pi * (U-0.5))
X1 = X1[(X1>-10) & (X1<10)]

plt.figure(1)
rv = scipy.stats.cauchy(loc = x0, scale = gamma)
plt.hist(X1, nbins, normed = True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw = 2)
plt.xlim((-10, 10))

plt.figure(2)
X2 = rv.rvs(N)
X2 = X2[(X2 > -10) & (X2 < 10)]
plt.hist(X2, nbins, normed = True)
x = np.linspace(-10, 10, 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw = 2 )
plt.xlim((-10, 10))
plt.show()
```

```python
import scipy.special
nbins = 50
k = 2.0
theta = 2.0

np.random.seed()
N= 100000
U = np.random.uniform(0.0, 1.0, N)

X1 = theta * scipy.special.gammaincinv(k, U)

plt.figure(1)
rv = scipy.stats.gamma(a = k, scale = theta)
plt.hist(X1, nbins, normed = True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw = 2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))
plt.figure(2)
X2 = rv.rvs(N)
plt.hist(X2, nbins, normed = True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw = 2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))

plt.show()
```
